```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel(r"C:\Users\matth\Downloads\healthcare_dataset proper.xlsx")
```

```python
#first few rows
data.head()
```

```python
#General info 
data.info()
```

```python
#Check for missing values
data.isnull().sum()
```

```python
#Top 10 most common medical conditions
common_conditions = data['Medical Condition'].value_counts().head(10)
print(common_conditions)
```

```python
#Graph for medical coonditions
plt.figure(figsize=(10, 5))
common_conditions.plot( kind = 'bar', color ='red')
plt.title('Top 10 Most Common Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.show()
```

```python
#Age based patterns on medical conditions
age_condition = data.groupby('Medical Condition')['Age'].mean().sort_values(ascending=True)
print(age_condition.head(10))
```

```python
plt.figure(figsize=(10, 5))
age_condition.head(10).plot(kind = 'bar', color ='orange')
plt.title(' Average Age by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Average Age')
plt.show()
```

```python
#Total Revenue per hospital
revenue_hospital = data.groupby('Hospital')['Billing Amount'].sum().sort_values(ascending=False)
print(revenue_hospital.head(10))
```

```python
#Visualization
plt.figure(figsize=(10, 5))
revenue_hospital.head(10).plot(kind = 'bar', color= 'green')
plt.title('Total Revenue By Hospital')
plt.xlabel('Hospital')
plt.ylabel('Total Billing Amount')
plt.show()
```

```python
#top 5 insurance providers based on patients
top_insurances = data['Insurance Provider'].value_counts().head(5)
print(top_insurances)
```

```python
#Visualization
plt.figure(figsize=(10, 5))
top_insurances.plot(kind= 'bar', color = 'red')
plt.title('Top 5 Insurance Providers')
plt.xlabel('Insurance Provider')
plt.ylabel('Number Of Patients')
plt.show()
```
